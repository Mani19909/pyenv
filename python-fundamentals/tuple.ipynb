Tuples (tuple)

Tuples are ordered, immutable sequences defined with parentheses (). Once created, their contents cannot be changed.

Characteristics and Use Cases

1. Ordered: items maintain position

2. Immutable: cannot add, remove, or change after creation

3. Useful for fixed records like coordinates, version numbers, or as dictionary keys


```python
host_port = ("127.0.0.1", 3000)
red_rgb = (255, 0, 0)
tuple_single_values = ("only-value",) # To create a single-item tuple, add a trailing comma
print(type(host_port))
print(type(tuple_single_values))

print(f"Host: {host_port[0]}")
print(red_rgb[-2:])
print(type(red_rgb[-2:]))

# host_port[0] = "192.168.1.1" # Uncommenting will raise a TypeError because tuples are immutable
```
